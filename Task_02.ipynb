```python
d2=pd.read_csv("test.csv")
for i in range (len(d2["Sex"])):
    if d2["Sex"][i]=='male':
        d2["Sex"][i]=1
    else :
        d2["Sex"][i]=0
d2=d2.drop(['Name'], axis=1)
d2=d2.drop(['Ticket'], axis=1)
d2=d2.drop(['PassengerId'], axis=1)
d2=d2.fillna(0)
for i in range (len(d2)):
    if d2["Cabin"][i]!=0:
        d2["Cabin"][i]=1
d2["Age"]=(d2["Age"]-np.min(d2["Age"]))/(np.max(d2["Age"])-np.min(d2["Age"]))
d2=d2.drop(["Embarked"],axis=1)
d2["Fare"]=(d2["Fare"]-np.min(d2["Fare"]))/(np.max(d2["Fare"])-np.min(d2["Fare"]))
y_test=d2["Survived"]
x_test=d2.drop(["Survived"],axis=1)
x_test=np.array(x_test).astype('float32')
y_test=np.array(y_test).astype('float32')

```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=0)
```

```python
model.fit(x_train,y_train)
```

```python
model_predict = model.predict(x_train)
```

```python
acc = accuracy_score(y_train,model_predict)
print(f"Accuracy Score:{acc}")
print("\n")
report = classification_report(y_train,model_predict)
print(f"Classification Report:{report}")
```

```python
y_model_predict = model.predict(x_test)
```

```python
y_acc = accuracy_score(y_test, y_model_predict)
print(f"Accuracy is:{y_acc}")

print("\n")

y_report = classification_report(y_test,y_model_predict)
print(f"Classification Report:{y_report}")
```
